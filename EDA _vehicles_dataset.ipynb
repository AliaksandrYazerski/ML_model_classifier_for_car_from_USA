### Research on dependence of price on car release year

```python
stats4 = df.groupby(['year'], as_index=False)[['price']].mean()
stats4 = stats4.rename(columns={'price': 'price_mean'})
stats4
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df.year, df.price)
plt.xlabel('Release year')
plt.ylabel('Price')
plt.title('Dependence of price on car release year')
plt.show();
```

```python
def get_out_boundaries(column):
    q25 = df[column].quantile(0.25)
    q75 = df[column].quantile(0.75)
    boundaries = (q25 - 1.5*(q75 - q25), q75 + 1.5*(q75 - q25))
    is_outlier = (df[column] < boundaries[0]) | (df[column] > boundaries[1])
    return boundaries, is_outlier
```

```python
boundaries_price, is_outlier_price = get_out_boundaries('price')
boundaries_price
```

```python
df[is_outlier_price]
```

```python
df.loc[is_outlier_price, 'price'] = boundaries_price[1]
```

```python
boundaries_year, is_outlier_year = get_out_boundaries('year')
boundaries_year
```

```python
df[is_outlier_year]
```

```python
df.loc[is_outlier_year, 'year'] = int(boundaries_year[0]) + 1
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df.year, df.price)
plt.xlabel('Release year')
plt.ylabel('Price')
plt.title('Dependence of price on car release year')
plt.show();
```

### Research on dependence of price category on car price and release year

```python
colors = {'low': 'green', 'medium': 'orange', 'high': 'red'}

fig, ax = plt.subplots(figsize=(12, 8))

for category in colors:
    color = colors[category]
    data = df[df.price_category == category]
    ax.hist(data.price, color=color, alpha=0.7)

plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Cars count, distributed by price, depending on car price category')
plt.legend(colors, title='Car price category');
```

```python
colors = {'low': 'green', 'medium': 'orange', 'high': 'red'}

fig, ax = plt.subplots(figsize=(12, 8))

for category in colors:
    color = colors[category]
    data = df[df.price_category == category]
    ax.scatter(data.year, data.price, color=color)

plt.xlabel('Release year')
plt.ylabel('Price')
plt.title('Car price distribution depending on release year, taking into account car price category')
plt.legend(colors, title='Car price category', loc='upper left');
```

## Correlations

```python
colors = {'low': 'green', 'medium': 'orange', 'high': 'red'}

fig, ax = plt.subplots(figsize=(12, 8))
for price_category in colors:
    color = colors[price_category]
    data = df[df['price_category'] == price_category]

    ax.hist(data['odometer'], bins=20, color=color, alpha=0.7)
    
plt.ylabel('Count')
plt.xlabel('Count of miles traveled')
plt.title('Count of miles traveled depending on car price category')
plt.legend(colors, title="Car price category");
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df['odometer'], df['year'])
plt.title("Count of miles traveled depending on car release year")
plt.xlabel('Count of miles traveled')
plt.ylabel('Release year')

plt.show();
```

```python
colors = {'low': 'green', 'medium': 'orange', 'high': 'red'}
fig, ax = plt.subplots(figsize=(12, 8))
for price_category in colors:
    color = colors[price_category]
    data = df[df['price_category'] == price_category]

    ax.scatter(data['odometer'], data['year'], c=color, label=price_category)

ax.legend(loc='upper right', title="Ценовая категория")
plt.title("Зависимость пробега от года выпуска автомобиля")
plt.xlabel('Пробег')
plt.ylabel('Год выпуска')
plt.show()
```
